# Covid 19 Analysis


```python
#importing lib
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing datasets
data= pd.read_csv('covid19_Confirmed_dataset.csv')
data.head()
```

```python
data.shape
```

```python
#deleting useless columns
df = data.drop(["Lat","Long"],axis=1, inplace= True)
```

```python
df=data.drop("Province/State",axis=1,inplace= True)
```

```python
#aggregating rows by countries
agg_data= data.groupby("Country/Region").sum()
agg_data.head(10)
```

```python
agg_data.shape
```

```python
#visualising data related to countries
agg_data.loc["India"].plot()
agg_data.loc["Finland"].plot()
agg_data.loc["Spain"].plot()

plt.legend()
```

```python
#finding maximum infection rate
agg_data.loc["Spain"].diff().max()
```

```python
agg_data.loc["Finland"].diff().max()
```

```python
agg_data.loc["Spain"].diff().max()
```

```python
countries = list(agg_data.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(agg_data.loc[c].diff().max())
agg_data["max_infection_rates"]= max_infection_rates
```

```python
agg_data
```

```python
corona_data=pd.DataFrame(agg_data["max_infection_rates"])
```
